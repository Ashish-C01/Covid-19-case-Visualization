```python
states=df2['State'].unique()
states=['Kerala', 'Delhi', 'Telangana', 'Rajasthan', 'Haryana',
 'Uttar Pradesh', 'Ladakh', 'Tamil Nadu', 'Jammu and Kashmir' ,'Karnataka'
 ,'Maharashtra', 'Punjab', 'Andhra Pradesh', 'Himachal Pradesh', 'Uttarakhand',
 'Odisha' ,'Puducherry', 'West Bengal', 'Chandigarh', 'Chhattisgarh' ,'Gujarat',
 'Madhya Pradesh', 'Bihar', 'Manipur', 'Goa', 'Mizoram',
 'Andaman and Nicobar Islands', 'Assam' ,'Jharkhand' ,'Arunachal Pradesh',
 'Nagaland', 'Tripura', 'Dadra and Nagar Haveli and Daman and Diu',
 'Meghalaya' ,'Sikkim','Lakshadweep']
lists =[[] for i in states]
k=0
for i in states:
    data=df2.groupby('State').get_group(i)
    for j in data.index:
        lists[k].append(data.loc[j])
    k=k+1
for i,j in enumerate(states):
    pd.Series(lists[i],name=j).to_csv(j+".csv")
```

```python
# 1- state 2-Total confirmed 'State', 'Confirmed', 'Recovered', 'Deceased', 'Other', 'Tested'
for i in states:
    confirmed=[]
    Recovered=[]
    Deceased=[]
    Date=[]
    data=df2.groupby('State').get_group(i)
    for j in data.index:
        confirmed.append(data.loc[j][2])
        Recovered.append(data.loc[j][3])
        Deceased.append(data.loc[j][4])
        Date.append(data.loc[j][0])
    prev_conf=confirmed[0]
    prev_rec=Recovered[0]
    prev_dec=Deceased[0]
    confirmed=pd.Series(confirmed).diff()
    Recovered=pd.Series(Recovered).diff()
    Deceased=pd.Series(Deceased).diff()
    confirmed[0]=prev_conf
    Recovered[0]=prev_rec
    Deceased[0]=prev_dec
    data=pd.DataFrame({
        'Confirmed': confirmed,
        'Recovered': Recovered,
        'Deceased': Deceased,
        'Date':Date
    })
    data.set_index('Date',inplace=True)
    data.to_csv(str(i)+".csv")
    data=pd.DataFrame()
```
